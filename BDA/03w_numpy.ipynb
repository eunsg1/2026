# numpy 주요 함수

```python
print(np.zeros(5))
print(np.zeros(5).dtype)
print(np.zeros(5, dtype = int))
print(np.zeros((2, 3)))
print(np.zeros(2, 3))
```

```python
print(np.full((2, 3), 10))
print(np.full([2, 3], [1, 2, 3]))
print(np.full([2, 3], [1, 2]))
```

```python
print(np.full([3, 4], range(4)))
```

```python
print(np.full([4, 3], fill_value = np.arange(4).reshape(4, 1)))
print(np.arange(4).reshape(4, 1))
```

```python
a = np.arange(6).reshape(3, 2)
b = np.array([[1], [2]])

print(a)
print(b)
print(a @ b)
```

```python
a = np.arange(1, 25).reshape(2, 3, 4)
print(a)


print(a[0, 2, 3])

print(a.ravel())
```

```python
a = np.arange(1, 4).reshape(1, 3)
b = np.arange(1, 4).reshape(3, 1)

print(a)
print(b)
print(a * b)
```

# 고급 색인

```python
x = np.arange(1, 11)
print(x)

print(x[1])
print(x[:5])
print(x[[2]])
print(x[[2, 3]])
print(x[[2, 3, 5, 8]])
print(x[np.array([5, 3, -4, 8])])
```

```python
x = np.arange(1, 57).reshape(7, 8)
print(x)
```

```python
print(x[0])
print(x[0, 1])
print(x[:, 2:8])
#x[, 2:8]

print(x[[0], [1]])
print(x[0, 1])

print(x[:, [5, 4]])
print(x[[1, 3]])
print(x[[1, 3], :])
```

```python
print(x[[1, 3], [2, 2]])
print(x[[1, 3], [2]])
print(x[[1, 3], 2])
```

```python
print(x[np.array([0, 2, 4]), 1])
print(x[[0, 2, 4], 1])
print(x[[0, 2, 4], :2])
```

```python
print(x[[0, 2, 4], [0, 1]])
print(x[np.array(0, 2, 4), np.array(0, 1)])
```

```python
a = np.arange(1, 41).reshape(5, 8)
print(a)
print(a[2:4, 3:7])
print(a[np.arange(2, 4)])
```

```python
print(a[np.arange(2, 4)][:, [1, 0]])
print(a[np.arange(2, 5)][:, [1, 0, 3, 5]])
print(a[np.arange(2, 5)])
print(a[np.arange(2, 5)][:, np.arange(-1, -8, -1)])
```
