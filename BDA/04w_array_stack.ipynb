```python
x = np.array([1, 2, 3])
y = np.array([10, 20, 30])
print(np.column_stack((x, y)))
print(np.column_stack((x[:, np.newaxis], y[:, np.newaxis])))

print(x[:, np.newaxis])
print(x.shape, x[:, np.newaxis].shape)
print(np.hstack((x, y)))
print(np.column_stack == np.hstack)
```

# 배열 결합 연습

```python
a = np.arange(1, 4)
b = np.arange(4, 7)
c = np.arange(7, 10)

print(np.vstack((a, b, c)))
```

```python
p = np.arange(1, 7).reshape(2, 3)
q = np.arange(11, 17).reshape(2, 3)
r = np.arange(21, 27).reshape(2, 3)

print(np.hstack((p, q, r)))
print(np.column_stack((p, q, r)))
print(np.concatenate((p, q, r)))
print(np.concatenate((p, q, r), axis = 1))
print(np.concatenate((p, q, r), axis = None))
print(np.concatenate((p, q, r), axis = -1))
```
