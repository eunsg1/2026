```python
n, bins, patches = plt.hist(iris['Sepal.Length'], bins = 20, color = 'yellowgreen', edgecolor = 'white')

print(n)
print(bins)
print(patches)
plt.show()
```

```python
n, bins, patches = plt.hist(iris['Sepal.Length'], bins = 20, color = 'yellowgreen', edgecolor = 'white')

for i in range(len(patches)):
  patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))

plt.show()
```

```python
import pandas as pd
pd.set_option('display.precision', 3)
corr_iris = iris[iris.columns[:-1].to_list()].corr()
corr_iris
```

```python
sns.heatmap(corr_iris, annot = True)
```

```python
sns.heatmap(corr_iris, annot = True, cmap = 'coolwarm', square = True, linewidths = 0.5, cbar_kws = {'shrink': 0.8})
```

```python
mask = np.zeros_like(corr_iris)
mask
```

```python
np.triu_indices_from(mask)
mask[np.triu_indices_from(mask)] = True
mask
```

```python
mask_new = mask[1:, :-1]
mask_new
```

```python
corr_iris_new = corr_iris.iloc[1:, :-1]
corr_iris_new
```

```python
sns.heatmap(corr_iris_new, annot = True, cmap = 'coolwarm', mask = mask_new, annot_kws = {"size": 10}, vmin = -1, vmax = 1, cbar_kws = {'shrink': .4}, linewidths = .5)
```
