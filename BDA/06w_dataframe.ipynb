```python
import pandas as pd

pf = pd.DataFrame(data = [[25, 35, 8, 18],
                         [18, 27, 10, 20],
                         [17, 17, 10, 19],
                         [12, 22, 9, 20],
                         [22, 34, 8, 16]],
                  index = ['윤일형', '강수희', '홍소희', '유한빈', '진수빈'],
                  columns = ['중간', '기말', '과제', '출석'])

print(pf)
```

```python
print(pf['중간'])
print(pf.중간)
print(pf['출석'])
print(pf[['출석']])
print(pf[['기말', '출석', '중간']])
print(pf[['기말']])
```

```python
print(pf['중간'] > 20)
print(pf[pf['중간'] > 20])
print(pf[pf['출석'] >= 19])
print(pf[(pf['출석'] >= 19) & (pf['기말'] >= 20)])
```

```python
print(pf.loc['윤일형'])
print(pf.loc[['윤일형']])
print(pf.loc[['홍소희', '윤일형']])
print(pf.loc['윤일형':'유한빈'])
print(pf.loc[pf.중간 > 20])
print(pf.loc[(pf.중간 > 10) & (pf.과제 == 10)])
```

```python
print(pf.loc[pf.index[0]])
print(pf.loc[[pf.index[0]]])
print(pf.loc[pf.index[1:4]])
print(pf.loc[pf.index[1:4]][['기말', '과제']])
```

```python
print(pf.loc[:, '과제'])
print(pf.loc[:, '중간':'과제'])
print(pf.loc[:, ['기말', '출석']])
```

```python
print(pf.loc[:, pf.columns[0]])
print(pf.loc[:, [pf.columns[0]]])
print(pf.loc[:, pf.columns[1:]])
print(pf.loc[:, pf.columns[1:3]])
print(pf.loc[:, pf.columns[1:]][['기말', '과제']])
```

```python
print(pf.loc['윤일형', '출석'])
print(pf.loc['윤일형':'유한빈', '기말'])
print(pf.loc['윤일형':'유한빈', '기말':'과제'])
print(pf.loc[pf.index[0], pf.columns[1]])
print(pf.loc[pf.index[0:3], pf.columns[2]])
print(pf.loc[pf.index[0:3], pf.columns[2:]])
```

```python
print(pf.iloc[1])
print(pf.iloc[[1]])
print(pf.iloc[[1, 3]])
print(pf.iloc[2:])
```

```python
print(pf.iloc[:, 2])
print(pf.iloc[:, [2]])
print(pf.iloc[:, 1:])
```

```python
print(pf.iloc[1, 1])
print(pf.iloc[1:3, 1:])
print(pf.iloc[:, 1:])
print(pf.iloc[[1, 3, 4], 1:])
print(pf.iloc[[1, 3, 4], [0, 1]])
```

```python
print(pf.at['윤일형', '과제'])
print(pf.at['유한빈', '기말'])
print(pf.at[pf.index[1], '출석'])
print(pf.at[pf.index[3], pf.columns[2]])
```

```python
print(pf.iat[1, 2])
print(pf.iat[2, 3])
```
