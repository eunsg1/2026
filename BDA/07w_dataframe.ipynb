```python
import pandas as pd

df = pd.read_csv('2016년졸음운전.csv', encoding = "euc-kr", index_col = 0)

print(df)
```

```python
print(df.describe())
print(df.T)
print(df.dtypes)
print(df.info())
```

```python
print(df.sum())
print(df.sum(axis = 0))
print(df.sum(axis = 1))
```

```python
print(df.cumsum(axis = 0))
print(df.cumsum(axis = 1))
```

```python
print(df.mean())
print(df.mean(axis = 0))
print(df.mean(axis = 1))
print(df.mean(axis = None))
```

```python
print(df.max())
print(df.max(axis = 0))
print(df.max(axis = 1))
print(df.max(axis = None))
print(df.min())
print(df.min(axis = 0))
print(df.min(axis = 1))
print(df.min(axis = None))
```

```python
print(df[['사망(명)', '부상(명)']].sum(axis = 1))
print(df.iloc[:, 1:].sum(axis = 1))
```

```python
df2 = df.copy()
df2['재해지수'] = df[['사망(명)', '부상(명)']].sum(axis = 1)
print(df2)
print(df2.sum())
df2.loc['총계'] = df2.sum(axis = 0)
print(df2)
```

```python
print(df2 > 10)
print(df2[df2 > 10])
print(df2[[True] * 6 + [False] * 7])
```

```python
print(df2['재해지수'] > 400)
print(df2[df2['재해지수'] > 400])
print(df['사망(명)'] == 7)
print(df['사망(명)'].isin([7]))
```

```python
print(df['사망(명)'].isin([7, 10]))
print(df[df['사망(명)'].isin([7, 10])])
print(df[(df['사망(명)'].isin([7, 10])) | (df['부상(명)'] > 350)])
print(df[df.index.str.contains('3')])
```

```python
df3 = df.copy()
print(df3)
df3[df3 > 10] = -df3
print(df3)
```

```python
df = pd.DataFrame({'A': range(1, 6),
                   'B': range(6, -3, -2),
                   'C': range(10, 5, -1),
                   'D clmns': range(3, -2, -1),
                   'E': ['python', 'java', 'python', 'go', 'python']})
print(df)
print(df.query('A > B'))
print(df[df.A > df.B])
```

```python
print(df.query('C == 10'))
print(df.query('A > 0'))
print(df.query('E == "python"'))
print(df.query('E != "python"'))
```

```python
print(df.query('A == `D clmns`'))
print(df.query('A == "D clmns"'))
print(df[df.A == df["D clmns"]])
print(df.query('A > 3 | B > 8'))
print(df.query('A > 2 & C < 8'))
print(df.query("E in ['python', 'go']"))
```

```python
num = 3
print(df.query('B > @num'))
print(df.query(f'B > {num}'))

s = "E.str.contains('py')"
print(df.query(s))
s = "E.str.contains('g')"
print(df.query(s))
s = "E.str.contains('on')"
print(df.query(s))
```
